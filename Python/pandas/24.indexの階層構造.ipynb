```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from numpy.random import randn
```

### 階層のindexにする

```python
ser1 = Series(randn(6), index=[[1,1,1,2,2,2], ['a', 'b', 'c', 'a', 'b', 'c']])
```

```python
ser1
```

```python
ser1.index
```

```python
ser1[1]
```

```python
ser1
```

```python
ser1[:, 'a']
```

### 階層indexのSeriesからDataFrameを作る

```python
dframe = ser1.unstack()
```

```python
dframe
```

```python
dframe.unstack()
```

```python
dframe.T.unstack()
```

### DataFrameから階層index

```python
dframe2 = DataFrame(np.arange(16).reshape((4 ,4)),
                   index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                   columns=[['NY', 'NY', 'LA', 'SF'], ['cold', 'hot', 'hot', 'cold']])
```

```python
dframe2
```

```python
dframe2.index.names = ['INDEX_1', 'INDEX_2']
```

```python
dframe2
```

```python
dframe2.columns.names = ['Cities', 'Temp']
```

```python
dframe2
```

```python
dframe2.swaplevel('Cities', 'Temp', axis=1)
```

```python
dframe2.sortlevel(1).sortlevel(0)
```

### levelごとの計算

```python
dframe2.sum(level='Temp', axis=1)
```
